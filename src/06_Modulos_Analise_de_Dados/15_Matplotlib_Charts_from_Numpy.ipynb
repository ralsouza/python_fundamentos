## Charts from Numpy

```
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```
# Create a scatter plot
plt.scatter(np.arange(50),np.random.randn(50))
plt.show()
```

```
# Create subplots
fig = plt.figure()

ax1 = fig.add_subplot(1,2,1)
ax1.plot(np.random.randn(50),color='r')

ax2 = fig.add_subplot(1,2,2)
ax2.scatter(np.arange(50),np.random.randn(50))
plt.show()
```

```
# Add many plots
_, ax = plt.subplots(2,3)

ax[0,1].plot(np.random.randn(50),color='g',linestyle='-')
ax[1,0].hist(np.random.randn(50))
ax[1,2].scatter(np.arange(50),np.random.randn(50),color='r')
plt.show()
```

```
# Axis controls
x = np.linspace(0,5,10)
y = x ** 2

fig, axes = plt.subplots(1,3,figsize=(12,4))

axes[0].plot(x, x**2, x, x**3)
axes[0].set_title('Axes with Default Range')

axes[1].plot(x,x**2,x,x**3)
axes[1].axis('tight')
axes[1].set_title('Smaller Axes')

axes[2].plot(x,x**2,x,x**3)
axes[2].set_ylim([0,60])
axes[2].set_xlim([2,5])
axes[2].set_title('Custom Axes')
```

```
# A thing that we need to consider, is the scales
# The scales can affect the chart interpretation
# We need to pay attention to correct scales
# Below the data are pretty much the same, but with different scales

fig, axes = plt.subplots(1,2,figsize=(10,4))

axes[0].plot(x, x**2, x, np.exp(x))
axes[0].set_title('Default Scale')

axes[1].plot(x, x**2, x, np.exp(x))
axes[1].set_yscale('log')
axes[1].set_title('Logaritimic Scale (y)')
```
