# Seaborn: Statistica Data Visualization
[http://seaborn.pydata.org](http://seaborn.pydata.org)

```
import seaborn as sea
import numpy as np
import pandas as pd
import random
import matplotlib.pyplot as plt
%matplotlib inline
```

```
# Load tips dataset from Seaborn
data = sea.load_dataset('tips')
data.head()
```

# Linear Regression is an equation to estimate a conditional (expected value) of an `y` variable. Given expected values of some others `x` variables. 

```
# The joinplot method create a plot of 2 variables with univariate and bivariate charts 
sea.jointplot('total_bill','tip',data=data,kind='reg')
```

```
# The lmplot() create a plot with data and regression models
sea.lmplot('total_bill','tip',data=data,col='smoker')
```

```
# Make an empty dataframe
df = pd.DataFrame()
```

```
# Fill the dataframe with random values
df['a'] = random.sample(range(1,100),25)
df['b'] = random.sample(range(1,100),25)
df.head()
```

```
# Plot a scatterplot
sea.lmplot('a','b',data=df,fit_reg=True)
```

```
# Density plot
sea.kdeplot(df.b)
```

```
sea.kdeplot(df.b,df.a)
```

```
sea.distplot(df.a)
```

```
# Histogram
plt.hist(df.a, alpha=.3)
sea.rugplot(df.a)
```

```
# Box plot
sea.boxplot([df.b,df.a])
```

```
# Violin plot
sea.violinplot([df.b,df.a])
```

```
# Heatmap
sea.heatmap([df.b,df.a],annot=True,fmt='d')
```

```
# Clustermap
sea.clustermap(df)
```
