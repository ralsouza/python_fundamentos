## Using random() method of Numpy


```
# Import packages
import numpy as np
import matplotlib.pyplot as plt 
%matplotlib inline
```

```
# Generate random numbers
rand_nums = np.random.randn(1000)

print(rand_nums)
```

```
# Create a histogram to show the distribuition of values
plt.show(plt.hist(rand_nums))
```

```
# Create a heat map
image = np.random.rand(30,30)
plt.imshow(image, cmap=plt.cm.hot)
plt.show(plt.colorbar())
```
