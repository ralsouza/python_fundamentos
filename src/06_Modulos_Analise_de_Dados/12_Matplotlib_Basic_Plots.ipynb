```
# The matplotlib.pyplot is a collection of function and styles that make works as the Matlab
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
```

```
mpl.__version__
```

```
# The .plot() method defines the chart axis
plt.plot([1,3,5],[2,5,7])
plt.show()
```

```
# Create axis data
x = [1,4,5]
y = [3,7,2]
```

```
# Add data into plot
plt.plot(x,y)

# Add labels to axis
plt.xlabel('Variable 1')
plt.ylabel('Variable 2')

# Add the title
plt.title('Test Plot')

# Showing the plot
plt.show()
```

```
x2, y2 = [1,2,3], [11,12,15]
```

```
# How to add a label
plt.plot(x2,y2, label='First Line')
plt.legend()
plt.show()
```
