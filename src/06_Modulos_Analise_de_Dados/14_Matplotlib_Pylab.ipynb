Pylab is a union between Pyplot from Matplotlib and Numpy

```
# Import libraries
import pylab as pl
%matplotlib inline
```

```
x = pl.linspace(0,5,10)
y = x ** 2

fig = pl.figure()

# Define the axes
axes = fig.add_axes([0.1,0.1,0.8,0.8])

axes.plot(x,y,'r')

axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('A Line Chart')
```

```
# Plots with two figures
x = pl.linspace(0,5,10)
y = x ** 2

fig = pl.figure()

# Axes of main figure
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
# Axes of secondary figure
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])

# Main figure
axes1.plot(x,y,'r')
axes1.set_xlabel('x')
axes1.set_ylabel('y')
axes1.set_title('Main Picture')

# Secondary figure
axes2.plot(y,x,'g')
axes2.set_xlabel('x')
axes2.set_ylabel('y')
axes2.set_title('Secondary Picture')
```

```
# Show chart side by side, usefull to comparing

# Define layout
fig, axes = pl.subplots(nrows=1,ncols=2)

for ax in axes:
  ax.plot(x,y,'r')
  ax.set_xlabel('foo')
  ax.set_ylabel('bar')
  ax.set_title('My Chart')

fig.tight_layout()
```
